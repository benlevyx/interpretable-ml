```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_silf(eps, beta):
    def silf(y):
        if 0 <= y and y <= (1 - beta) * eps:
            return 0
        elif (1 - beta) * eps <= y and y <= (1 + beta) * eps:
            return ((y - (1 - beta) * eps) ** 2) / (4 * beta * eps)
        elif y >= (1 + beta) * eps:
            return y - eps
        else:
            raise ValueError("y cannot be negative")
    return np.vectorize(silf)
```

```python
eps = 3
beta = 1.1
silf = make_silf(eps, beta)
```

```python
xx = np.linspace(0, 50, 1000)
yy = silf(xx)
```

```python
plt.plot(xx, yy)
plt.plot(xx, xx, linestyle='--', c='r')
```
